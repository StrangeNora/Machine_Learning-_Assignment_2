```python
import seaborn as sns

popularity_df = df[(df[['Stream', 'Likes', 'Comments', 'Views']] > 0).all(axis=1)].copy()

popularity_df['Popularity'] = (
    popularity_df['Stream'].rank(pct=True) +
    popularity_df['Likes'].rank(pct=True) +
    popularity_df['Comments'].rank(pct=True) +
    popularity_df['Views'].rank(pct=True)
) / 4

# Replot the 2D density plot with no warning
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=popularity_df,
    x='Valence',
    y='Energy',
    weights=popularity_df['Popularity'],
    fill=True,
    cmap='viridis',
    thresh=0.01,
    levels=100
)
plt.title("Valence vs Energy (Weighted by Popularity)")
plt.xlabel("Valence")
plt.ylabel("Energy")
plt.tight_layout()
plt.show()
```

this plot shows where the most popular songs are concentrated based on:

- Valence (musical positivity)

- Energy (intensity and activity)

The brighter areas represent combinations that correlate with higher popularity, based on a normalized blend of streams, likes, comments, and views.

### Part B:

```python
df['Album_type_clean'] = df['Album_type'].str.lower().replace('compilation', 'album')

loudness_min = df['Loudness'].min()

loudness_max = df['Loudness'].max()

normalized_loudness = (df['Loudness'] - loudness_min) / (loudness_max - loudness_min)

def classify_pitch(key):
    if key in [0, 1, 2, 3]:  # C, C#, D, D#
        return 'low'
    elif key in [4, 5, 6, 7]:  # E, F, F#, G
        return 'medium'
    else:  # G#, A, A#, B
        return 'high'

df = df.assign(
    Album_Song_Count = df.groupby(['Album_type_clean', 'Album'])['Track'].transform('count'),
    Average_Artist_Song_Views = df.groupby('Artist')['Views'].transform('mean'),
    Song_Name_Length = df['Track'].astype(str).apply(lambda x: len(x.split())),
    Total_Album_Length = df.groupby(['Album_type_clean', 'Album'])['Duration_ms'].transform('sum'),
    Fitness_for_Clubs = (df['Danceability'] + df['Energy'] + df['Valence'] + normalized_loudness) / 4,
    Pitch = df['Key'].apply(classify_pitch),
    Popular_Site = df.apply(lambda x: 'youtube' if x['Views'] > x['Stream'] else 'spotify', axis=1),
    Is_Remix = df[['Title', 'Track', 'Description']].astype(str).apply(lambda x: 'remix' in ' '.join(x).lower(), axis=1),
    Streams_per_Minute = df['Stream'] / (df['Duration_ms'] / 60000),
    Likes_to_Views = df['Likes'] / df['Views']
)

df['Mood_Label'] = pd.cut(
    df['Energy'] * df['Danceability'],
    bins=[-float('inf'), 0.2, 0.4, 0.6, 0.8, float('inf')],
    labels=['Mellow', 'Chill', 'Groovy', 'Upbeat', 'Hype']
)

df.drop(columns='Album_type_clean', inplace=True)
```

Besides the required features, I created six custom ones to better capture the musical qualities, listener behavior, and engagement patterns for each song:

- Pitch
This groups the song’s Key into ‘low’, ‘medium’, or ‘high’ pitch ranges. Since pitch can shape a song’s mood and emotional feel, it might also relate to genre or what listeners prefer.

- Popular_Site
A simple yes/no feature that shows whether a song is more popular on YouTube or Spotify by comparing Views and Streams. This helps us understand where the audience is most active, which could guide platform-specific promotion.

- Is_Remix
This flags whether a song is likely a remix by checking if the word “remix” appears in its Title, Track, or Description. Remixes often behave differently in terms of audience and streaming trends compared to original songs.

- Streams per Minute
Calculated by dividing total streams by the song length in minutes, this feature shows how often a song gets played per minute. It’s a good way to spot songs with high replay value or viral popularity.

- Likes_to_Views_Ratio
This measures YouTube engagement by dividing Likes by Views. A high ratio means listeners really like the song, while a low ratio could mean people are less engaged or indifferent.

- Mood_Label
Based on how energetic and danceable a song is, this groups tracks into five moods: ‘Mellow’, ‘Chill’, ‘Groovy’, ‘Upbeat’, and ‘Hype’. It helps capture the overall vibe, which is great for mood-based recommendations or playlists.

```python
import numpy as np

df['Comments'] = df['Comments'].fillna(0)
df['Likes'] = df['Likes'].fillna(0)
df['Views'] = df['Views'].fillna(0)
```

We imputed missing values in the Comments column with 0, assuming that missing values likely represent songs where comments weren’t available or disabled. This preserves all rows for modeling and prevents errors in algorithms that don’t support NaN values. It's a safe and logical default for this type of engagement metric.

```python
df['Log_Views'] = np.log1p(df['Views'])
df['Log_Streams'] = np.log1p(df['Stream'])
df['Log_Likes'] = np.log1p(df['Likes'])
```

We used a log transformation (specifically log1p, which is safe for zero values) on the features Views, Likes, and Streams because these numbers can vary wildly — from very small to extremely large. Applying the log helps to:

Smooth out big differences in scale

Make the data more “normal” or balanced

Tone down the effect of really extreme values

This kind of adjustment often helps models work better, especially those that rely on distances or assume data is roughly normally distributed.

```python
df.drop(columns=['Uri', 'Url_spotify', 'Url_youtube', 'Unnamed: 0'], inplace=True)
```

We removed features like URLs and long text fields (Uri, Url_spotify, Url_youtube) because they don’t really help with making predictions. These kinds of data can add unnecessary noise or even cause the model to overfit. Plus, they’re tricky to use directly without extra processing like natural language techniques. By dropping them, we keep the dataset simpler and focus on the useful, structured information the model can learn from.
